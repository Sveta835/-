```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f (x):
  return 3*x**2
```

```python
def df (x):
  return 6*x
```

```python
x0 =5
x1 = x0 - l*df(x0)
l = 0.1
res = {}
```

```python
while abs(x0 - x1) > 0.00000001:
  res.update({x0:f(x0)})
  x0 = x1
  x1 = x0 - l*df(x0)
```

```python
data = pd.DataFrame(list(res.items()), columns=['x', 'y'])
data
```

1. Метод градиентного спуска для одной переменной

```python
x = np.linspace(-5, 5, 400)
y = 3*x**2
plt.plot(x, y, color = 'blue')
plt.plot(data['x'], data['y'], color = 'red')
plt.show()
```

2. Метод градиентного спуска для двух переменных

```python
def f (x, y):
  return 3*x**2 + 5*y**2
```

```python
def df (x,y):
  return np.array([6*x, 10*y])
```

```python
x_y_0 =np.array([5, 10])

x_y_1 = x_y_0 - l*df(x_y_0[0], x_y_0[1])
l =0.1
res = pd.DataFrame(columns = ['x', 'y', 'f'])
```

```python
i = 0
while np.linalg.norm(x_y_0 - x_y_1) > 0.00000001:
  res.loc[i, 'x'] = x_y_0[0]
  res.loc[i, 'y'] = x_y_0[0]
  res.loc[i, 'f'] = f(x_y_0[0], x_y_0[1])
  x_y_0 = x_y_1
  x_y_1 = x_y_0 - l*df(x_y_0[0], x_y_0[1])
  i += 1
res
```
