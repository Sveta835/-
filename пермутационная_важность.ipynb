```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set();

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_validate
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/katarina74/ml_lessons/main/lesson_2/data/techparams_train.csv")
X = df.drop(["target", "index"], axis=1)
y = df[["target"]]
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
reg = LinearRegression().fit(train_X, train_y)
```

```python
mse_0 = mean_squared_error(reg.predict(test_X), test_y)
```

```python
r_0 = reg.score(test_X, test_y)
```

```python
dict_1 = {}
```

```python
for x in train_X.columns:
  train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.33, random_state=42)
  train_X_1 = train_X
  train_X_1[x] = np.random.permutation(train_X_1[x].values)
  reg_1 = LinearRegression().fit(train_X_1, train_y)
  dict_1[x] = (mean_squared_error(reg_1.predict(test_X), test_y) - mse_0)
sorted(dict_1.items(), key=lambda item: item[1], reverse=True)
```

```python
data = pd.DataFrame(dict_1.items(), columns=['feature', 'test'])

plt.figure(figsize =(25, 6))
sns.barplot(data = data.sort_values(by = 'test').tail(10), x = 'feature', y = 'test')
plt.grid()
plt.show()
```

```python
dict_2 = {}
```

```python
for x in train_X.columns:
  train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.33, random_state=42)
  train_X_1 = train_X
  train_X_1[x] = np.random.permutation(train_X_1[x].values)
  reg_1 = LinearRegression().fit(train_X_1, train_y)
  dict_2[x] = (r_0 - reg_1.score(test_X, test_y))
sorted(dict_2.items(), key=lambda item: item[1], reverse=True)
```
